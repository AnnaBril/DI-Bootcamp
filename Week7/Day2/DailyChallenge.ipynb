```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
import re
from dateutil import parser
```

### Scraping the Webpage

```python
driver = webdriver.Chrome()

url = "https://www.bbc.com/weather/293397"
driver.get(url)

# Allow time for the page to load (you may need to adjust the sleep duration)
time.sleep(5)

# Extract the HTML content after dynamic content is loaded
html_content = driver.page_source

# Close the browser
driver.quit()

# Use BeautifulSoup to parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Extract relevant data (e.g., dates, numerical values, categories)
dates = [date.text.strip() for date in soup.select('.wr-date__long')]
temperatures = [temp.text.strip() for temp in soup.find_all('span', class_='wr-value--temperature--c')]
descriptions = [desc.text.strip() for desc in soup.find_all('div', class_='wr-day__weather-type-description')]

# Check the lengths of the lists
print("Number of Dates:", len(dates))
print("Number of Temperatures:", len(temperatures))
print("Number of Descriptions:", len(descriptions))

# Find the common length among the lists
common_length = min(len(dates), len(temperatures), len(descriptions))

# Organize the scraped data into a structured format (e.g., Pandas DataFrame)
data = {'Date': dates[:common_length], 'Temperature (°C)': temperatures[:common_length], 'Description': descriptions[:common_length]}
ta_weather = pd.DataFrame(data)

ta_weather
```

```python
ta_weather.info()
```

### Clean and preprocess the data

```python
df_copy = ta_weather.copy()
```

```python
# Convert the "Temperature" column to integer
df_copy['Temperature (°C)'] = df_copy['Temperature (°C)'].str.rstrip('°').astype(int)
```

```python
df_copy['Date'] = df_copy['Date'].apply(lambda x: re.sub(r'\b(?:Monday|Tuesday|Wednesday|Thursday|Friday|Saturday|Sunday)\b\s*', '', x))
df_copy['Date'] = df_copy['Date'].apply(lambda x: re.sub(r'(st|nd|rd|th)', '', x))
df_copy['Date'] = df_copy['Date'].str.strip() + ' 2024'
```

```python
df_copy['Date'] = '2024 ' + df_copy['Date']

# Convert the "Date" column to datetime
df_copy['Date'] = pd.to_datetime(df_copy['Date'], format='%Y %d %B', errors='coerce')
df_copy
```

### Data Analysis

```python
#obtaining basic information about a numeric column
df_copy['Temperature (°C)'].describe()
```

```python
rainy_days = df_copy[df_copy['Description'].str.contains('rain', case=False)].shape[0]
print(f"Total rainy days: {rainy_days}")

sunny_days = df_copy[df_copy['Description'].str.contains('sun', case=False)].shape[0]
print(f"Total sunny days: {sunny_days}")
```

### Data Visualization

```python
sns.set(style="whitegrid")
# Line plot for temperature trends over time
plt.figure(figsize=(11, 5))
sns.lineplot(x='Date', y='Temperature (°C)', data=df_copy, marker='o')
plt.title('Temperature trends over time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart for the distribution of weather descriptions
plt.figure(figsize=(10, 6))
sns.countplot(x='Description', data=df_copy, order=df_copy['Description'].value_counts().index, palette='viridis')
plt.title('Distribution of weather descriptions')
plt.xlabel('Weather description')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```
