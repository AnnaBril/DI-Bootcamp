```python
#  Converting categorical data into a format suitable for analysis, such as one-hot encoding
df_encoded = pd.get_dummies(df_cleaned, columns=['extended', 'crit1', 'success', 'suicide'])
df_encoded.head(10)
```

```python
# Creating an instance of MinMaxScaler
scaler = MinMaxScaler()

# Applying normalization to numeric columns
df_cleaned['nkill_normalized'] = scaler.fit_transform(df_cleaned[['nkill']])
df_cleaned['nkillus_normalized'] = scaler.fit_transform(df_cleaned[['nkillus']])
df_cleaned['nkillter_normalized'] = scaler.fit_transform(df_cleaned[['nkillter']])

# Creating an instance of StandardScaler
scaler1 = StandardScaler()

# Applying standardization to numeric columns
df_cleaned['nwound_standardized'] = scaler1.fit_transform(df_cleaned[['nwound']])
df_cleaned['nwoundus_standardized'] = scaler1.fit_transform(df_cleaned[['nwoundus']])
df_cleaned['nwoundte_standardized'] = scaler1.fit_transform(df_cleaned[['nwoundte']])
df_cleaned.head(10)
```

## 3. Exploratory Data Analysis:
Data Exploration: Conduct exploratory data analysis on the cleaned and integrated dataset.
Visualization: Create visualizations to reveal patterns and insights related to terrorism events in their socio-economic context.

```python
attacks_year = pd.pivot_table(df_cleaned, values='eventid', index='iyear', aggfunc='count').reset_index()
attacks_year = attacks_year.rename(columns={'iyear': 'year', 'eventid': 'count'})
attacks_year
```

```python
# A timeline showing the distribution of terrorist acts by year
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")
sns.lineplot(x='year', y='count', data=attacks_year, marker='o')
plt.title('Number of terrorist attacks over years')
plt.xlabel('Year')
plt.ylabel('Number of attacks')
plt.show()
```

```python
# Building a map of terrorist attacks with cleared coordinates
Q1_lat = df_cleaned['latitude'].quantile(0.25)
Q3_lat = df_cleaned['latitude'].quantile(0.75)
IQR_lat = Q3_lat - Q1_lat

Q1_lon = df_cleaned['longitude'].quantile(0.25)
Q3_lon = df_cleaned['longitude'].quantile(0.75)
IQR_lon = Q3_lon - Q1_lon

lower_bound_lat = Q1_lat - 1.5 * IQR_lat
upper_bound_lat = Q3_lat + 1.5 * IQR_lat

lower_bound_lon = Q1_lon - 1.5 * IQR_lon
upper_bound_lon = Q3_lon + 1.5 * IQR_lon

df_cleaned_coordinates = df_cleaned[(df_cleaned['latitude'] >= lower_bound_lat) & (df_cleaned['latitude'] <= upper_bound_lat) &
                            (df_cleaned['longitude'] >= lower_bound_lon) & (df_cleaned['longitude'] <= upper_bound_lon)]

plt.figure(figsize=(12, 8))
sns.scatterplot(x='longitude', y='latitude', data=df_cleaned_coordinates, hue='region_txt', legend='full', s=50)
plt.title('Global distribution of terrorist attacks')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```

```python
# the distribution of different types of terrorist attacks
plt.figure(figsize=(14, 8))
sns.countplot(x='attacktype1_txt', data=df_cleaned, order=df_cleaned['attacktype1_txt'].value_counts().index)
plt.title('Distribution of terrorist attack types')
plt.xlabel('Attack type')
plt.ylabel('Number of attacks')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# the distribution of different types of targets
plt.figure(figsize=(14, 8))
sns.countplot(x='targtype1_txt', data=df_cleaned, order=df_cleaned['targtype1_txt'].value_counts().index)
plt.title('Distribution of targets types')
plt.xlabel('Target type')
plt.ylabel('Number of attacks')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# the distribution of terrorist attacks by country
# Getting the top 30 countries by number of attacks
top_countries = df_cleaned['country_txt'].value_counts().nlargest(30).index

# DataFrame filtering for top 30 countries only
df_top_countries = df_cleaned[df_cleaned['country_txt'].isin(top_countries)]

# Plotting a graph
plt.figure(figsize=(14, 10))
sns.countplot(y='country_txt', data=df_top_countries, order=top_countries)
plt.title('Distribution of terrorist attacks by country (Top 30)')
plt.xlabel('Number of attacks')
plt.ylabel('Country')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# histograms of the distribution of categorical columns, where 0 is no, 1 is yes
for i, column in enumerate(['extended', 'success', 'suicide', 'individual']):
    sns.countplot(x=column, data=df_cleaned, ax=axes[i // 2, i % 2])
    axes[i // 2, i % 2].set_title(f'Distribution of {column}')
    axes[i // 2, i % 2].set_xlabel(column)
    axes[i // 2, i % 2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

- The original dataset had many columns with a lot of missing values. After cleaning there were 53 columns left.
