##  Exercise 1 : Basic Usage Of SciPy
Instructions
Import the SciPy library

```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

##  Exercise 2 : Descriptive Statistics
Instructions
Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.

```python
# Sample dataset
data = [12, 15, 13, 12, 18, 20, 22, 21]

# Calculate descriptive statistics using SciPy
result = stats.describe(data)

mean = result.mean
variance = result.variance
std_dev = result.variance**0.5

print("Mean:", mean)
print("Variance:", variance)
print("Standard deviation:", std_dev)
```

##  Exercise 3 : Understanding Distributions
Instructions
Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.

```python
mean = 50
std_dev = 10

# Generate random data following a normal distribution
data1 = np.random.normal(loc=mean, scale=std_dev, size=1000)

# Plot the histogram of the generated data
plt.hist(data1, bins=30, density=True, alpha=0.7, color='lightgreen', label='Generated data')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution')

plt.title('Normal distribution with mean=50, std dev=10')
plt.xlabel('Value')
plt.ylabel('Probability density')
plt.legend()

plt.show()
```

##  Exercise 4 : T-Test Application
Instructions
Perform a T-test on two sets of randomly generated data. Hint: Use numpy.random.normal() to generate the datasets.

```python
# Set random seed for reproducibility
np.random.seed(0)

# Generate two sets of random data with different means
data2 = np.random.normal(loc=50, scale=10, size=100)
data3 = np.random.normal(loc=55, scale=10, size=100)

# Perform an independent two-sample t-test
t_statistic, p_value = stats.ttest_ind(data2, data3)

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the two groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the two groups.")
```
