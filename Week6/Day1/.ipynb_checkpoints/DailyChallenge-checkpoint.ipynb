### 1. Data Preparation:

Hint 1: Use np.random.uniform(low, high, size) to generate the temperature data.
Hint 2: Create a DataFrame using pd.DataFrame(data, index, columns) with appropriate index and columns.

Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.

Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set seed for reproducibility
np.random.seed(42)

# Generate synthetic temperature data
temperature_data = np.random.uniform(low=-5, high=35, size=(10, 12))

cities = ["City" + str(i) for i in range(1, 11)]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Create a DataFrame with city names as index and months as columns
weather = pd.DataFrame(temperature_data, index=cities, columns=months)
weather
```

### 2. Data Analysis:

Hint 1: Calculate the annual average temperature using DataFrame.mean(axis).
Hint 2: Find the city with the highest and lowest average temperature using idxmax() and idxmin() methods.

Calculate the annual average temperature for each city.

Identify the city with the highest and lowest average temperature for the year.

```python
# Calculate the annual average temperature for each city
annual_avg_temperatures = weather.mean(axis=1)

city_highest_avg = annual_avg_temperatures.idxmax()
city_lowest_avg = annual_avg_temperatures.idxmin()

print("Annual average temperatures:")
print(annual_avg_temperatures)
print("\nCity with the highest average temperature:", city_highest_avg)
print("City with the lowest average temperature:", city_lowest_avg)
```

### 3. Data Visualization:

Hint 1: Use plt.plot() for each city’s temperature trends. Adjust the linewidth and linestyle for the highlighted cities.
Hint 2: Use plt.title(), plt.xlabel(), plt.ylabel(), and plt.legend() to add details to your plot.

Use Matplotlib to create a line plot showing the monthly temperature trends for each city.

Highlight the city with the highest and lowest average annual temperature in a different color or line style.

```python
# Plot monthly temperature trends for each city
for city in weather.index:
    if city == city_highest_avg:
        plt.plot(months, weather.loc[city], label=city, linewidth=2, linestyle='dashed', marker='o')
    elif city == city_lowest_avg:
        plt.plot(months, weather.loc[city], label=city, linewidth=2, linestyle='dotted', marker='s')
    else:
        plt.plot(months, weather.loc[city], label=city, linestyle='solid', marker='.')

plt.title('Monthly temperature trends for each city')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.legend()

plt.show()
```
