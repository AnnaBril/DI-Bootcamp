## Preparing for the study

```python
import pandas as pd
import requests
import json
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import re
import ast
from IPython.display import Image
from IPython.display import display, HTML
# libraries import
```

```python
url = "https://swapi.dev/api/planets/?format=json"
response = requests.get(url)

#  Convert the response to JSON format
data = response.json()

# Extracting the list of planets from the data
planets = data.get("results", [])

# Empty list to store data for each planet
data_list = []

# Cycle to process each planet
for planet in planets:
    # Create a URL for each planet
    planet_url = planet["url"]
    planet_response = requests.get(planet_url)

    # Convert the response for each planet into JSON format
    planet_data = planet_response.json()

    # Resident Processing
    residents_list = []
    for resident_url in planet_data.get("residents", []):
        resident_response = requests.get(resident_url)
        resident_data = resident_response.json()
        residents_list.append(resident_data)

    # Replace references to residents with their data
    planet_data["residents"] = residents_list
    
     # Movie processing
    films_list = []
    for film_url in planet_data.get("films", []):
        film_response = requests.get(film_url)
        film_data = film_response.json()
        films_list.append(film_data)
        
     # Replace links to movies with their data
    planet_data["films"] = films_list

    # Adding data to the list
    data_list.append(planet_data)

# Create a dataframe from the data list
df = pd.DataFrame(data_list)
```

## Data preprocessing

```python
# Delete "created", "edited", "url" columns
df = df.drop(["created", "edited", "url"], axis=1)
```

```python
# Convert columns to integer type
df["rotation_period"] = pd.to_numeric(df["rotation_period"]).astype('Int64')
df["orbital_period"] = pd.to_numeric(df["orbital_period"]).astype('Int64')
df["diameter"] = pd.to_numeric(df["diameter"]).astype('Int64')
df.info()
```

```python
# Convert gravity column to float type
df["gravity"] = df["gravity"].replace("N/A", None)
df["gravity"] = df["gravity"].apply(lambda x: float(re.search(r'(\d+(\.\d+)?)', str(x)).group(1)) if pd.notnull(x) else None)
```

```python
df["surface_water"] = df["surface_water"].replace("unknown", 0)
# Convert column surface_water to integer type
df["surface_water"] = pd.to_numeric(df["surface_water"]).astype('Int64')
```

```python
df["population"] = df["population"].replace("unknown", 0)
# Convert column population to integer type
df["population"] = pd.to_numeric(df["population"]).astype('Int64')
```

```python
# Making the values in the terrain column consistent
df['terrain'] = df['terrain'].replace({
    'jungle, rainforests': 'jungles, rainforests',
    'tundra, ice caves, mountain ranges': 'tundra, ice caves, mountains',
    'swamp, jungles': 'swamps, jungles',
    'grassy hills, swamps, forests, mountains': 'grasslands, swamps, forests, mountains'
})
```

## Selecting the table with residents from the source dataset

```python
# Create an empty list for rows with information about residents
residents_data = []

# Iterate over the rows of the source DataFrame
for index, row in df.iterrows():
    # Iterate through the list of residents
    for resident in row['residents']:
        # Create a dictionary with resident data
        resident_data = {
            'planet_name': row['name'],  
            'resident_name': resident['name'],
            'resident_height': resident['height'],
            'resident_mass': resident['mass'],
            'resident_hair_color': resident['hair_color'],
            'resident_skin_color': resident['skin_color'],
            'resident_eye_color': resident['eye_color'],
            'resident_birth_year': resident['birth_year'],
            'resident_gender': resident['gender'],
            'resident_homeworld': resident['homeworld'],
            'resident_films': ', '.join(resident['films']),
            'resident_species': ', '.join(resident['species']),
            'resident_vehicles': ', '.join(resident['vehicles']),
            'resident_starships': ', '.join(resident['starships']),
            'resident_created': resident['created'],
            'resident_edited': resident['edited'],
            'resident_url': resident['url']
        }
        residents_data.append(resident_data)

# Create a new DataFrame from the list of dictionaries
residents_df = pd.DataFrame(residents_data)

residents_df.head(20)
```

## Data analysis and visualization of a table with data on planets

```python
numeric_columns = df.select_dtypes(include='number').columns

# Building histograms with KDE for each numeric column
for column in numeric_columns:
    sns.histplot(df[column], bins=20, kde=True)
    plt.title(f'Distribution of {column}')
    plt.grid(True)
    plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    The values in the rotation_period column are most often 23 and 24. Most values in the orbital_period column are in the range of 300 to 600. Most values in the diameter column are in the range of 7000 to 12000. Most values in the surface_water column are in the range of 0 to 15%. In the population column there is a planet Coruscant, which has tens of times more population than the others, and because of this the graph is uninformative.

```python
correlation_matrix = df.corr()

# building a correlation matrix using seaborn
plt.figure(figsize=(9, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation matrix')
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    The strongest correlation was found between planet diameter and gravity (0.94), between diameter and rotation period negative correlation -0.78, between diameter and orbital period 0.68, between gravity and rotation period negative correlation -0.67, between gravity and orbital period 0.66. The weakest correlation was found between gravity and water surface - negative correlation -0.08, planet diameter and population 0.1, gravity and population 0.1.

```python
# building boxplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 12))
fig.suptitle('Boxplots for numeric columns', fontsize=16)

axes = axes.flatten()

for i, col in enumerate(df.select_dtypes(include=['int64', 'float64']).columns):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(f'Boxplot for {col}')
    axes[i].grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    Many columns contain outliers - values that are very different from the rest. Because of this, boxplots on these data are uninformative.

```python
# Value limits for removing abnormal values from the graph
value_limits = {
    'orbital_period': (300, 600),
    'diameter': (7000, 20000),
    'population': (0, 4500000000) 
}

# Constructing a pairplot with value constraints
pairplot = sns.pairplot(df, kind='scatter', corner=True)

for i, var in enumerate(pairplot.data.columns):
    if var in value_limits:
        pairplot.data[var] = pairplot.data[var].clip(*value_limits[var])

plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    In general, the scatter plots confirm the relationships identified in the correlation matrix plot.
