Content
The happiness scores and rankings use data from the Gallup World Poll. The scores are based on answers to the main life evaluation question asked in the poll. This question, known as the Cantril ladder, asks respondents to think of a ladder with the best possible life for them being a 10 and the worst possible life being a 0 and to rate their own current lives on that scale. The scores are from nationally representative samples for the years 2013-2016 and use the Gallup weights to make the estimates representative. The columns following the happiness score estimate the extent to which each of six factors – economic production, social support, life expectancy, freedom, absence of corruption, and generosity – contribute to making life evaluations higher in each country than they are in Dystopia, a hypothetical country that has values equal to the world’s lowest national averages for each of the six factors. They have no impact on the total score reported for each country, but they do explain why some countries rank higher than others.
    It contains 9 columns and 156 rows. Only column "Country or region" has an object type, "Overall rank" - integer, the rest are float.
    
    There are no duplicates and missing values in the dataset.

### Social Support and Happiness:

Employ a Matplotlib scatter plot to investigate the relationship between ‘Social support’ and ‘Score’.

```python
plt.figure(figsize=(9, 5))
plt.scatter(world_happiness['Social support'], world_happiness['Score'], color='blue', alpha=0.7)
plt.title('Relationship between social support and happiness score')
plt.xlabel('Social support')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    In general, there is a tendency to increase the score with an increase in the value of Social support.

### Regional Comparisons of GDP and Health:

Create a Matplotlib subplot that compares ‘GDP per Capita’ and ‘Healthy Life Expectancy’ across different regions.

```python
fig, ax = plt.subplots()
ax.scatter(world_happiness['GDP per capita'], world_happiness['Healthy life expectancy'], color='green')

ax.set_title('Comparison of GDP per capita and Healthy life expectancy across different regions')
ax.set_xlabel('GDP per capita')
ax.set_ylabel('Healthy life expectancy')
plt.grid(True)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
index = range(len(world_happiness['Country or region']))

line1, = ax.plot(index, world_happiness['GDP per capita'], label='GDP per capita', marker='o')
line2, = ax.plot(index, world_happiness['Healthy life expectancy'], label='Healthy life expectancy', marker='o')

ax.set_title('Comparison of GDP per capita and Healthy life expectancy across all countries')
ax.set_xlabel('Country')
ax.set_ylabel('Values')
ax.set_xticks(index[::4])
ax.set_xticklabels(world_happiness['Country or region'][::4], rotation=45, ha='right')
ax.legend()

plt.tight_layout() 
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    In general, the graphs show a relationship between GDP per capita and Healthy life expectancy. In most cases, with a high GDP per capita, the Healthy life expectancy indicator is also high.

### Regional Comparisons of GDP and Health:
Use a bar plot for ‘GDP per Capita’ and a line plot for ‘Healthy Life Expectancy’ on the same axes to observe how economic strength relates to health outcomes in different regions.

```python
fig, ax = plt.subplots(figsize=(11, 7))

# Bar plot for 'GDP per Capita'
bar_width = 0.4
index = np.arange(len(world_happiness['Country or region']))
bars_gdp = ax.bar(index, world_happiness['GDP per capita'], width=bar_width, label='GDP per capita', color='skyblue')

# Line plot for 'Healthy Life Expectancy'
line_health, = ax.plot(index, world_happiness['Healthy life expectancy'], label='Healthy life expectancy', color='red', marker='o')

ax.set_title('Comparison of GDP per capita and Healthy life expectancy across all countries')
ax.set_xlabel('Country')
ax.set_ylabel('Values')
ax.set_xticks(index[::4])
ax.set_xticklabels(world_happiness['Country or region'][::4], rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```
