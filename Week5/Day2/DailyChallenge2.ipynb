# Daily Challenge: Analyzing The IMDB Movie Dataset With Matplotlib And Seaborn

## Load and Inspect Data:

Use pandas to load the IMDB Movie Dataset.
Inspect the data to understand its structure and contents.

```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#import plotly.express as px
```

```python
imdb_top = pd.read_csv('imdb_top_1000.csv')

imdb_top.info()
imdb_top.head(10)
```

```python
# determining the percentage of missing values in each column
pd.DataFrame(round(imdb_top.isna().mean().sort_values(ascending=False)*100,1)).style.background_gradient('coolwarm')
```

```python
# checking for duplicate strings
imdb_top.duplicated().sum()
```

<div style="border:solid green 2px; padding: 20px">
    IMDB Dataset of top 1000 movies and tv shows.
    It contains 16 columns and 1000 rows.
   
   **This dataset contains the following data:**
- Poster_Link - Link of the poster that imdb using
- Series_Title = Name of the movie
- Released_Year - Year at which that movie released
- Certificate - Certificate earned by that movie
- Runtime - Total runtime of the movie
- Genre - Genre of the movie
- IMDB_Rating - Rating of the movie at IMDB site
- Overview - mini story/ summary
- Meta_score - Score earned by the movie
- Director - Name of the Director
- Star1,Star2,Star3,Star4 - Name of the Stars
- No_of_votes - Total number of votes
- Gross - Money earned by that movie
    
The missing values are in the columns "Gross", "Meta_score" and "Certificate". There are no duplicates in the dataset. The format of the "Released_Year", "Runtime" and "Gross" column values is object.

## Data Preprocessing:
Handle missing values and convert data types as needed.

```python
# converting a column type to a date and float
imdb_top['Released_Year'] = pd.to_datetime(imdb_top['Released_Year'], format='%Y', errors='coerce')
#imdb_top['Gross'] = pd.to_numeric(imdb_top['Gross'], errors='coerce')
imdb_top['Certificate'].fillna('not known', inplace=True)
```

```python
# Convert to float
imdb_top['Gross'] = imdb_top['Gross'].str.replace(',', '')
imdb_top['Gross'] = pd.to_numeric(imdb_top['Gross'])
imdb_top.info()
```

## Rating Trends Over Years:
Create a line plot with Matplotlib showing trends in IMDB_Rating over the years.

```python
imdb_top['IMDB_Rating'].describe()
```

```python
# Group by year and calculate the mean of IMDB_Rating
rating_trends = imdb_top.groupby('Released_Year')['IMDB_Rating'].mean().reset_index()

# Create a line plot
plt.figure(figsize=(11, 6))
plt.plot(rating_trends['Released_Year'], rating_trends['IMDB_Rating'], marker='o', linestyle='-')
plt.title('IMDB rating trends over the years')
plt.xlabel('Year')
plt.ylabel('Average IMDB Rating')
plt.grid(True)
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    The maximum average rating for films was 8.5 in 1936. The minimum rating in 1935 was less than 7.8. Sharp jumps in the average rating were also noticeable in 1943, 1957, and 1994. In the period from 1995 to 2019, approximately the same values of the average rating are observed - in the range from 7.87 to 8.02. And in 2020, another leap is noticeable.

## Genre Popularity Analysis:
Generate a Seaborn bar plot to compare the number of movies across different genres.

```python
# Split the genres into separate rows
df_genres = imdb_top['Genre'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).rename('Genre')

imdb_top = imdb_top.drop('Genre', axis=1).join(df_genres)
genre_counts = imdb_top['Genre'].value_counts()

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='magma')
plt.title('Number of movies across different genres')
plt.xlabel('Genre')
plt.ylabel('Number of movies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    The leader in the number of movies in the genre is drama. There are more than three times as many such movies as there are movies of other genres. Next, there are almost the same number of movies in the genres of comedy, crime, adventure. Least of all musical movies and westerns.

## Director’s Impact on Ratings:
Use a Matplotlib scatter plot to examine the relationship between directors and movie ratings.

```python
director_ratings = imdb_top.groupby('Director')['IMDB_Rating'].mean().reset_index().sort_values(by='IMDB_Rating', ascending=False).round(2)
director_ratings.head(20)
```

```python
# Create a scatter plot
plt.figure(figsize=(12, 7))
plt.scatter(imdb_top['Director'], imdb_top['IMDB_Rating'], alpha=0.5, edgecolors='w', s=100)

plt.xticks([])
plt.title('Director’s impact on ratings')
plt.xlabel('Director')
plt.ylabel('IMDB Rating')
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    In most cases, a director can have movies with completely different ratings, and the more movies there are, the more likely the rating values will be scattered.

## Star Power Analysis:
Employ a Seaborn pairplot to explore relationships involving Stars and their impact on IMDB_Rating and Gross.

```python
sns.pairplot(imdb_top[['IMDB_Rating', 'Gross', 'Star1', 'Star2', 'Star3', 'Star4']])
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    As with directors, famous actors are not always the key to a successful movie.

## Box Plot of Genres vs. Ratings:
Visualize the distribution of IMDB_Rating across different Genres using a Seaborn box plot.

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Genre', y='IMDB_Rating', data=imdb_top)
plt.xticks(rotation=45, ha='right') 
plt.title('Distribution of IMDB ratings across genres')
plt.xlabel('Genre')
plt.ylabel('IMDB Rating')
plt.ylim(7.58, 9.05)
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    Western movie have the largest interquartile range and higher ratings. There are quite a few outliers in terms of rating value for all genres. Musical movies and film-noir have the least spread of ratings.

## Correlation Heatmap:

Create a heatmap using Seaborn to identify correlations between numerical columns like IMDB_Rating, Meta_score, No_of_votes, and Gross.

```python
correlation_matrix = imdb_top[['IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Gross']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation heatmap')
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    The largest correlation between the "No_of_Votes" and "Gross" columns is 0.59. Also, the correlation between the columns "No_of_Votes" and "IMDB_Rating" is 0.54. The weakest correlation between the columns "Gross" and "Meta_score" is -0.1, as well as between "Meta_score" and "No_of_Votes" is 0.02.
